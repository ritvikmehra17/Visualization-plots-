```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
weather=pd.read_csv("Rainier_Weather.csv")
```

```python
import plotly.express as px
```

```python
import plotly.express as px

fig = px.scatter(weather, x="Temperature AVG",color="Date",
           color_continuous_scale=px.colors.sequential.Viridis, render_mode="webgl")
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig=make_subplots(specs=[[{'secondary_y':True}]])
fig.add_trace(go.Scatter(x=weather["Temperature AVG"].head(),
                        y=weather["Battery Voltage AVG"].head()),secondary_y=False)
fig.add_trace(go.Scatter(x=weather["Wind Direction AVG"].head(),
                        y=weather["Solare Radiation AVG"].head()),secondary_y=True)

fig.update_layout(title_text="Double y axis")
fig.update_yaxes(title_text="asdad",secondary_y=False)
fig.update_yaxes(title_text="zxczxc",secondary_y=True)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig=make_subplots(rows=1,cols=2,specs=[[{"secondary_y":True},{"secondary_y":True}]])


fig.add_trace(go.Scatter(x=weather["Wind Direction AVG"].head(),
                        y=weather["Wind Speed Daily AVG"].head(),name="yaxis1"),
              row=1,col=1,secondary_y=False)
fig.add_trace(go.Scatter(x=weather["Battery Voltage AVG"].head(),
                        y=weather["Solare Radiation AVG"].head()),
             row=1,col=1,secondary_y=True)

fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[2, 52, 62], name="yaxis5 data"),
    row=1, col=2, secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[40, 50, 60], name="yaxis6 data"),
    row=1, col=2, secondary_y=True,
)
fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=weather["Temperature AVG"].head(),
    y=weather["Relative Humidity AVG"].head(),
    name="yaxis1 data"
))


fig.add_trace(go.Scatter(
    x=weather["Battery Voltage AVG"].head(),
    y=weather["Solare Radiation AVG"].head(),
    name="yaxis2 data",
    yaxis="y2"
))

fig.add_trace(go.Scatter(
    x=weather["Wind Direction AVG"].head(),
    y=weather["Wind Speed Daily AVG"].head(),
    name="yaxis3 data",
    yaxis="y3"
))

fig.add_trace(go.Scatter(
    x=weather["Relative Humidity AVG"].head(),
    y=weather["Solare Radiation AVG"].head(),
    name="yaxis4 data",
    yaxis="y4"
))


# Create axis objects
fig.update_layout(
    xaxis=dict(
        domain=[0.3, 0.7]
    ),
    yaxis=dict(
        title="yaxis title",
        titlefont=dict(
            color="#1f77b4"
        ),
        tickfont=dict(
            color="#1f77b4"
        )
    ),
    yaxis2=dict(
        title="yaxis2 title",
        titlefont=dict(
            color="#ff7f0e"
        ),
        tickfont=dict(
            color="#ff7f0e"
        ),
        anchor="free",
        overlaying="y",
        side="left",
        position=0.15
    ),
    yaxis3=dict(
        title="yaxis4 title",
        titlefont=dict(
            color="#d62728"
        ),
        tickfont=dict(
            color="#d62728"
        ),
        anchor="x",
        overlaying="y",
        side="right"
    ),
    yaxis4=dict(
        title="yaxis5 title",
        titlefont=dict(
            color="#9467bd"
        ),
        tickfont=dict(
            color="#9467bd"
        ),
        anchor="free",
        overlaying="y",
        side="right",
        position=0.85
    )
)

# Update layout properties
fig.update_layout(
    title_text="multiple y-axes example",
    width=800,
)

fig.show()
```
