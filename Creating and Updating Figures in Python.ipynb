```python
import plotly.graph_objects as go
fig = go.Figure(
    data=[go.Bar(x=weather["Date"].head(), y=weather["Solare Radiation AVG"].head())],
    layout=go.Layout(
        title=go.layout.Title(text="A solare radiation Chart")
    )
)
fig.show()
```

```python
import plotly.express as px
fig=px.scatter(weather,x="Solare Radiation AVG",y="Temperature AVG",color="Date")
fig.show()
```

```python
from plotly.subplots import make_subplots
```

```python
fig=make_subplots(rows=2,cols=2)
fig.add_trace(go.Bar(y=weather["Wind Speed Daily AVG"].head()),row=1,col=1)
fig.add_trace(go.Scatter(y=weather["Wind Speed Daily AVG"].head(),mode="lines"),
              row=1,col=2)
fig.add_trace(go.Scatter(y=weather["Wind Speed Daily AVG"].head()),row=2,col=1)
fig.show()
```

```python
import plotly.express as px
fig=px.scatter(weather,x="Solare Radiation AVG",y="Temperature AVG",color="Date")
fig.add_trace(go.Scatter(x=weather["Solare Radiation AVG"].head(),
                        y=weather["Temperature AVG"].head(),
                        line=go.scatter.Line(),
                        showlegend=False))
fig.show()
```
